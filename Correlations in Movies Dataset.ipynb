```python
#import libraries

import pandas as pd
import seaborn as sns
import  numpy as np

import matplotlib
import matplotlib.pyplot as plt

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)  #now this adjusts the size of the plots that  we will create

#now we read this data and save it in a dataframe (df)

df = pd.read_csv(r'C:\Users\aullah010\Desktop\archive\movies.csv')
```

```python
#look at the  data

df.head()
```

```python
#checking missing data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print ('{} --- {}%'.format(col,percent_missing))
```

```python
#look for data types of our columns

df.dtypes
```

```python
#changing datatype

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#create correct year column

df['yearcorrect'] = df['released'].astype(str).str[:4]
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
pd.set_option('display.max_rows',None)
```

```python
#now we set out to find  corellations between:
# 1.budget vs gross revenue
# 2.Company vs gross revenue 
#we find correlations by building scatter plots

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')
plt.show()
```

```python
#using seaborn to create a regression plot

sns.regplot(x='budget',y='gross',data=df ,scatter_kws={"color":"blue"}, line_kws={"color":"red"})
```

```python

#fundamentally there are many types of correlations
#Kendall, pearson, spearman
#now we're looking at correlations
#default is pearson

df.corr()
#df.corr(method='pearson')
#df.corr(method='spearman')
#df.corr(method='kendall')

```

```python
#high correlation between budget and gross
#high correlation between votes and gross

#now we will visualize this correlation matrix
correlation_matrix = df.corr(method='pearson')

#sns= seaborn lib
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
#lighter colors -->higher correlation
#darker colors --> weaker correlation

    
```

```python
#we want to include  company name in our correlation matrix, but since  its a string and not a number so we cannot add it
#so now we will create numbers against each company and then plot them to correlation matrix

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'): #type object = string in python
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
df_numerized.corr()
```

```python
correlation_matrix_numerized = df_numerized.corr()
correlation_pairs = correlation_matrix_numerized.unstack()
correlation_pairs
```

```python
sorted_pairs = correlation_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5 ]
high_corr
```
